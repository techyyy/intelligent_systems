<h1>Gnome sorting(natural order)</h1>

```python
def gnome_sort(arr):
    index = 0
    n = len(arr)
    
    while index < n:
        if index == 0 or arr[index - 1] <= arr[index]:
            index += 1
        else:
            arr[index], arr[index - 1] = arr[index - 1], arr[index]
            index -= 1   
    return arr

def task_10():
    user_list = [78, 34, 23, 7, 5, 3, 3]
    sorted_list = gnome_sort(user_list)
    print("Sorted list:", sorted_list)
```
